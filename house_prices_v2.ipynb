```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
X_train = pd.read_csv('/kaggle/input/romanatural0/train.csv')
X_test = pd.read_csv('/kaggle/input/romanatural0/test.csv')
y_train = X_train['SalePrice']
X_train = X_train.fillna(0)
X_test = X_test.fillna(0)

X_train = X_train.drop(columns=['SalePrice'])

cat_mask = (X_train.dtypes == "object").values
X_train_cat = X_train.loc[:, cat_mask]
X_test_cat = X_test.loc[:, cat_mask]
X_train_no = X_train.loc[:, ~cat_mask]
X_test_no = X_test.loc[:, ~cat_mask]
comb = pd.concat([X_train_cat, X_test_cat], axis=0)
comb_dummies = pd.get_dummies(comb, drop_first=True)

train_size = len(X_train_cat)
train = comb_dummies[:train_size]  
test = comb_dummies[train_size:]
train = train.replace({True: 1, False: 0})
test = test.replace({True: 1, False: 0})
train_concat = pd.concat([train, X_train_no], axis=1)
test_concat = pd.concat([test, X_test_no], axis=1)


fromthebox = LinearRegression()
w = fromthebox.fit(train_concat, y_train)
y_test = fromthebox.predict(test_concat)
output = pd.DataFrame({'Id': X_test.Id, 'SalePrice': y_test})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```
